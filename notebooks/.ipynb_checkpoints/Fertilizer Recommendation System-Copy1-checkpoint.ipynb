## Preprocessing using One-Hot Encoder

```python
y = fertilizer_df['Fertilizer Name'].copy()
X = fertilizer_df.drop('Fertilizer Name',axis=1).copy()
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3,4])],remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
X
```

## Train-test split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,shuffle=True,random_state=42)
```

## Feature Scaling

```python
sc = StandardScaler()
print(X_train)
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Machine Learning Models

### Random Forest

```python
RFClassifier = RandomForestClassifier(n_estimators=200,criterion='gini',random_state=42)
RFClassifier.fit(X_train,y_train)
```

```python
y_pred = RFClassifier.predict(X_test)
score = accuracy_score(y_test,y_pred)
print("Random Forest Classifier's Accuracy is: ",score*100,"%")
print("\nClassification Report: \n\n",classification_report(y_test,y_pred))
```

```python
cross_score = cross_val_score(RFClassifier, X_test,y_test,cv=5).mean()
print("Cross Validation Score of Random Forest Classifier: ",cross_score*100)
```

### eXtreme Gradient Boosting (XGBoost)

```python
le = LabelEncoder()
XGBoostClassifier = XGBClassifier()
XGBoostClassifier.fit(X_train,y_train)
```

```python
y_pred = XGBoostClassifier.predict(X_test)
score = accuracy_score(y_test,y_pred)
print("XGBoost Classifier's Accuracy is: ",score*100,"%")
print("\nClassification Report: \n\n",classification_report(y_test,y_pred))
```

```python
cross_score = cross_val_score(XGBoostClassifier, X_test,y_test,cv=5).mean()
print("Cross Validation Score of XGBoost Classifier: ",cross_score*100,"%")
```

### LightGBM

```python
LightGBMClassifier = LGBMClassifier()
LightGBMClassifier.fit(X_train,y_train)
```

```python
y_pred = LightGBMClassifier.predict(X_test)
score = accuracy_score(y_test,y_pred)
print("LightGBM Classifier's Accuracy is: ",score*100,"%")
print("\nClassification Report: \n\n",classification_report(y_test,y_pred))
```

```python
cross_score = cross_val_score(LightGBMClassifier, X_test,y_test,cv=5).mean()
print("Cross Validation Score of LightGBM Classifier: ",cross_score*100,"%")
```

### Stacking Classifier

```python
classifierfinal = StackingClassifier(
    classifiers=[
        RandomForestClassifier(),
        XGBClassifier(),
        LGBMClassifier()
    ],
    use_probas=True,
    meta_classifier=XGBClassifier()
)

classifierfinal.fit(X_train,y_train)
pred_stack3 = classifierfinal.predict(X_test)

print(classification_report(y_test,pred_stack3))
```

```python
cross_score = cross_val_score(classifierfinal, X_test,y_test,cv=5).mean()
print("Cross Validation Score of Stacking Classifier: ",cross_score*100,"%")
```

```python
import lightgbm as lgb

yc = fertilizer_df['Fertilizer Name'].copy()
Xc = fertilizer_df.copy().drop('Fertilizer Name',axis=1)

X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(Xc, yc, test_size=0.2, random_state=42)

# params = {
#     'boosting_type': 'gbdt',
#     'objective': 'binary',
#     'metric': 'binary_logloss',
#     'num_leaves': 31,
#     'learning_rate': 0.05,
#     'feature_fraction': 0.9,
#     'bagging_fraction': 0.8,
#     'bagging_freq': 5,
#     'lambda_l1': 0.1,  # L1 regularization parameter
#     'lambda_l2': 0.1,  # L2 regularization parameter
#     'verbose': -1
#     }

params = {
    'objective': 'multiclass',
    'num_classes': 7,
    'metric': 'multi_logloss',
    'boosting_type': 'gbdt',
    'learning_rate': 0.05,
    'num_leaves': 31,
    'max_depth': -1,
    'min_child_samples': 20,
    'subsample': 0.8,
     'lambda_l1': 0.1,  # L1 regularization parameter
    'lambda_l2': 0.1,  # L2 regularization parameter
    'colsample_bytree': 0.8,
    'reg_alpha': 0.1,  # Regularization parameter
}

# new_lgb = LGBMClassifier()
train_data = lgb.Dataset(X_train_c, label=y_train_c)

# Train the model with regularization
model = lgb.train(params, train_data, num_boost_round=100, valid_sets=[train_data], early_stopping_rounds=10)

y_pred_c = model.predict(X_test_c)
print("-----------predicted values--------",y_pred_c)
y_pred_c = [np.argmax(line) for line in y_pred_c]
accuracy = accuracy_score(y_test_c, y_pred_c)

print("-------- actual values--------",y_test_c)
print("-----------predicted values--------",y_pred_c)
print(f'Accuracy: {accuracy}')
# cv_results = lgb.cv(params, train_data, num_boost_round=100, nfold=5, metrics='multi_logloss', stratified=True, shuffle=True, early_stopping_rounds=10, verbose_eval=False, callbacks=[lgb.callback.print_evaluation(show_stdv=False), lgb.callback.early_stopping(10)])
# print('Regularized cross-validation logloss:', cv_results['multi_logloss-mean'][-1])
# cross_score = cross_val_score(model, X_test_c,y_test_c,cv=5).mean()
# print("Cross Validation Score of Stacking Classifier: ",cross_score*100,"%")
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import lightgbm as lgb

def rebalance_dataset(X_train,y_train):
    sm = SMOTE(random_state=42)
    X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
    return X_train_res,y_train_res

def evaluate_model(model,X,y):
    y_predict = model.predict(X)
    print(y_predict)
    score = accuracy_score(y, y_predict.round())
    print("accuracy: ", score)
#     print("Recall: ",recall_score(y_test, y_test_predict))
#     print("f1-score: ",f1_score(y_test, y_test_predict))
    
def detect_drift(model,prev_score,X,y):
    y_predict = model.predict(X)
    updated_score = accuracy_score(y,y_predict)
    drift = updated_score - prev_score
    print("drift",drift,updated_score,prev_score)
    if drift > 0.1:
        return True
    else:
        return False

def add_regularization(X,y):
    
    params = {
    'boosting_type': 'gbdt',
    'objective': 'binary',
    'metric': 'binary_logloss',
    'num_leaves': 31,
    'learning_rate': 0.05,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'lambda_l1': 0.1,  # L1 regularization parameter
    'lambda_l2': 0.1,  # L2 regularization parameter
    'verbose': -1
    }

    # Create the dataset object
    train_data = lgb.Dataset(X, label=y)

    # Train the model with regularization
    model = lgb.train(params, train_data, num_boost_round=100, valid_sets=[train_data], early_stopping_rounds=10)
    
    return model
    
```
